```python

full_df_aps = full_df[full_df['image']=='all']

# full_df_aps["arc_threshold"]  = full_df_aps["line_threshold"] + 5
full_df_aps = full_df_aps[['AP_line', 'AP_circle', 'AP_arc',
       'line_threshold', 'circle_threshold','arc_threshold', 'model', 'epoch', 'image']]
full_df_aps['AP_average'] = full_df_aps[['AP_line', 'AP_circle', 'AP_arc']].mean(axis=1)
# Get a list of all unique models

models = full_df_aps['model'].unique()
# models = ['full', 'no-dn-no-2stage']
# Create a colormap
cmap = cm.get_cmap('tab10')
marker_dict = {'line': 'x', 'circle': 'o', 'arc': '^',  'average': 's'}

color_dict = {model: cmap(i) for i, model in enumerate(models)}
# Increase the size of the subplots
# Increase the size of the subplots
fig, axs = plt.subplots(1, 3, figsize=(24, 10))  # Adjust the figsize parameter as needed

for i, primitive in enumerate(['line', 'circle', 'arc']):
    ax = axs[i]
    for model_name in models: 
        if 'LETR' in model_name and primitive == 'arc':
            continue
        df = full_df_aps[(full_df_aps['model'] == model_name)]

        ap_primitive = f'AP_{primitive}'
        threshold_primitive = f'{primitive}_threshold' if primitive != 'average' else 'line_threshold'
        
        ax.scatter(df[threshold_primitive], df[ap_primitive], color=color_dict[model_name], marker=marker_dict[primitive], s=120)
        # Increase the line width
        ax.plot(sorted(df[threshold_primitive]),sorted(df[ap_primitive]), color=color_dict[model_name], label=model_name, linewidth=3)

    # Increase the font size of the labels
    ax.set_ylabel(f'AP {primitive}', fontsize=46)
    ax.set_xlabel('Threshold distance $\delta_{max}$', fontsize=46)

    ax.xaxis.set_label_coords(0.5, -0.15)  # Adjust the label position
    ax.grid(True)
    # Set yscale to log for line and circle plots
    if primitive in ['line', 'circle', 'arc']:
        ax.set_xscale('log')
    ax.set_ylim(0, 0.95)
    ax.set_xticks(ticks = df[threshold_primitive], labels = df[threshold_primitive], rotation=45, fontsize=30)
    yticks = [ 0.2, 0.4,  0.6, 0.8]
    ax.set_yticks(ticks = yticks, labels = yticks, fontsize=30)
    # Create the legend patches
legend_patches = [mpatches.Patch(color=color, label=model) for model, color in color_dict.items() ] 

# Increase the font size of the legend
fig.legend(handles=legend_patches, fontsize=30, bbox_to_anchor=(1.0, 1.0) ,loc='upper left')

plt.tight_layout()
fig.savefig('ap_thresholds.pdf', bbox_inches='tight', format='pdf')
```
